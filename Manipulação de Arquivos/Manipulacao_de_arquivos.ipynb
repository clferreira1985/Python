```python
import pandas as pd
import sqlite3
```

```python
#Open engine connection

con=sqlite3.connect ('/content/cdd.db')

# Create a cursor object

cur = con.cursor()
```

```python
# Criar uma Consulta: Query

Query = cur.execute('select * from produtos')

# Salvando o resultado da query em um dataframe: df ( Save results of the query to DataFrame: df)
df = pd.DataFrame(Query.fetchall(), columns=['ID', 'DT', 'PRD', 'VL'])

# Fechando a Conex√£o (Close connection)

con.commit()

# Print head of Dataframe df
df.head()
```
